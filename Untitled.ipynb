```julia
using BenchmarkTools
using Serialization
using Plots
using Statistics
using Combinatorics
```

```julia
SymCPU_benchmarks = open(deserialize, "SymCPU_benchmarks.jls")
SymGPU_benchmarks = open(deserialize, "SymGPU_benchmarks.jls")
LogSymCPU_benchmarks = open(deserialize, "LogSymCPU_benchmarks.jls")
LogSymGPU_benchmarks = open(deserialize, "LogSymGPU_benchmarks.jls")
```

```julia
SymCPU_benchmarks
```

```julia
player_range = 5:5:50
action_range = 2:6

times = zeros(Float64, size(SymCPU_benchmarks))

for (a,action) in enumerate(action_range)
    for (p,player) in enumerate(player_range)
        try
            times[p,a] = round(mean(SymCPU_benchmarks[p,a].times) / 1e9, digits=4)
        catch e
        end
    end
end

plot(player_range, times, lw=3)
```
