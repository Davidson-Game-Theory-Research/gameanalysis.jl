```julia
plot()
profiles = Matrix{Float64}(undef,2,201)
profiles[1,:] .= 0:.1:20
profiles[2,:] .= 20:-.1:0
for i in 1:10
    h = random_gaussian(Dirichlet([1,1]), mean_scale=20)
    plot!(0:.1:20, [h(profiles[:,i]) for i in 1:201])
end
plot!()
```
